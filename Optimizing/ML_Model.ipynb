```python
with open('fillers.txt', 'r', encoding='utf-8') as file:
    data = file.read()

print(f"Number of tokens (words): {len(data.split())}")
print(data[:500]) 

from nltk.tokenize import sent_tokenize


filler_sents = sent_tokenize(data)

print(filler_sents)
```

```python
filler_starts = [
    "yeah", "ok", "okay", "uh", "um", "ah", "hmm", "right", "alright", 
    "sure", "got it", "absolutely", "indeed", "well", "you know", "i mean","cool", 'welcome','hello','bye',
    "family","thank you", "thanks", "no problem", "no worries", "youre welcome", 
    "hi", "hey", "hello again", "good morning", "good afternoon", "good evening", "bye", "goodbye",
    "maybe", "sort of", "kind of", "I guess", "possibly", "probably", "hopefully", "I suppose", "I think", "I feel like",
    "like", "so", "well", "you know", "I mean", "basically", "actually", "literally", "seriously", "anyway", "just",
    "how are you", "hows it going", "whats up", "long time no see", "nice to see you", "how was your weekend"


]

def filter_fillers(sentences):
    cleaned = []

    for i, s in enumerate(sentences):
        words = s.split()
        num_words = len(words)
        s_lower = s.lower().strip()

        prev_short = i > 0 and len(sentences[i-1].split()) < 5
        next_short = i < len(sentences)-1 and len(sentences[i+1].split()) < 5
        if num_words < 5 and (prev_short or next_short):
            continue

        if num_words <= 3 and any(s_lower == f for f in filler_starts):
            continue

        if num_words < 5 and any(s_lower.startswith(f) for f in filler_starts):
            continue

        cleaned.append(s)

    return cleaned

cleaned_meet_sentences = filter_fillers(meet_sents)
cleaned_filler_sentences = filter_fillers(filler_sents)

print(f"Original meet sentences: {len(meet_sents)}")
print(f"Original filler sentences: {len(filler_sents)}")

print(f"After filler removal: (meet) {len(cleaned_meet_sentences)}")
print(f"After filler removal: (filler){len(cleaned_filler_sentences)}")

removed = [s for s in meet_sents if s not in cleaned_meet_sentences]
print("Sample removed sentences:", removed)
```

```python
from nltk.tokenize import word_tokenize
total_words = 0
for sentence in cleaned_meet_sentences:
    tokens = word_tokenize(sentence)
    total_words += len(tokens)

print(f"Total number of words: {total_words}")

total_words = 0
for sentence in cleaned_filler_sentences:
    tokens = word_tokenize(sentence)
    total_words += len(tokens)

print(f"Total number of words: {total_words}")
```

```python
import pandas as pd
import random


meet_data = [(s, 0) for s in cleaned_meet_sentences]  # label 0 for meeting
filler_data = [(s, 1) for s in cleaned_filler_sentences]  # label 1 for filler

combined_data = meet_data + filler_data

random.shuffle(combined_data)

df = pd.DataFrame(combined_data, columns=["sentence", "label"])

print(df.head())
```

```python
filler_words = [
    "um", "uh", "like", "you know", "i mean", "so", "well", "actually", "basically",
    "literally", "sort of", "kind of", "you see", "right", "hmm"
]
def count_filler_words(sentence, filler_list):
    sentence = sentence.lower()
    count = 0
    for filler in filler_list:
        count += sentence.count(filler)
    return count
```

```python
df['filler_count'] = df['sentence'].apply(lambda x: count_filler_words(x, filler_words))
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
filler_phrases = [
    "um", "uh", "like", "you know", "i mean", "so", "well", 
    "actually", "basically", "literally", "sort of", "kind of"
]

filler_reference_text = " ".join(filler_phrases)
filler_embedding = model.encode(filler_reference_text)
sentence_embeddings = np.array(df['embedding'].tolist())

similarities = cosine_similarity(sentence_embeddings, [filler_embedding]).flatten()

df['filler_similarity_score'] = similarities
```

```python
df['word_count'] = df['sentence'].apply(lambda x: len(x.split()))
df['char_count'] = df['sentence'].apply(len)
df['avg_word_length'] = df['char_count'] / df['word_count']
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

df['stopword_ratio'] = df['sentence'].apply(
    lambda x: sum(1 for word in x.lower().split() if word in stop_words) / len(x.split())
)
```

```python
filler_starts = ('so', 'well', 'actually', 'like', 'i mean', 'you know')

df['starts_with_filler'] = df['sentence'].str.lower().apply(
    lambda x: any(x.startswith(f) for f in filler_starts)
).astype(int)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")  # Lightweight English model

def count_pos_tags(sentence, pos_tag):
    doc = nlp(sentence)
    return sum(1 for token in doc if token.pos_ == pos_tag)
```

```python
df['interjection_count'] = df['sentence'].apply(lambda x: count_pos_tags(x, 'INTJ'))
df['adverb_count'] = df['sentence'].apply(lambda x: count_pos_tags(x, 'ADV'))
```

```python
import numpy as np
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

# Download required resources if not already
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger_eng')
# nltk.download('stopwords')

# Load sentence transformer
embedder = SentenceTransformer('all-MiniLM-L6-v2',device = 'cpu')

# Define your filler words (expand if needed)
filler_words = ['um', 'uh', 'like', 'you know', 'actually', 'so', 'basically', 'well']
 
# Optional: Precompute filler sentence embeddings
filler_sentences = [
    "um", "uh", "you know", "like", "i mean", "so", "actually", "basically", "well"
]
filler_embeddings = embedder.encode(filler_sentences)
avg_filler_embedding = np.mean(filler_embeddings, axis=0).reshape(1, -1)

# Define handcrafted feature function
def handcrafted_feature_fn(sentence, filler_words):
    words = word_tokenize(sentence.lower())
    stop_words = set(stopwords.words('english'))

    filler_count = sum(word in filler_words for word in words)
    word_count = len(words)
    char_count = len(sentence)
    avg_word_len = np.mean([len(word) for word in words]) if words else 0
    stopword_ratio = sum(word in stop_words for word in words) / word_count if word_count else 0
    starts_with_filler = int(words[0] in filler_words) if words else 0
    pos_tags = [tag for word, tag in pos_tag(words)]
    interjection_count = pos_tags.count('UH')
    adverb_count = sum(tag in ['RB', 'RBR', 'RBS'] for tag in pos_tags)

    # Compute filler similarity score
    sentence_embedding = embedder.encode([sentence])[0].reshape(1, -1)
    similarity = cosine_similarity(sentence_embedding, avg_filler_embedding)[0][0]

    return [
        filler_count,
        similarity,
        word_count,
        char_count,
        avg_word_len,
        stopword_ratio,
        starts_with_filler,
        interjection_count,
        adverb_count
    ]

# Prediction function
def predict_sentence(sentence, model):
    # Get embedding
    embedding = embedder.encode([sentence])[0]  # (384,)

    # Get handcrafted features
    handcrafted = handcrafted_feature_fn(sentence, filler_words)  # (9,)

    # Combine into a single input vector
    x_input = np.hstack([embedding, handcrafted]).reshape(1, -1)  # (1, 393)

    # Predict
    label = model.predict(x_input)[0]
    prob = model.predict_proba(x_input)[0]

    return label, prob
```

```python
# Test sentences with their true labels (1 = filler, 0 = not filler)
test_sentences = [
    # Filler (1)
    ("Actually, I'm not sure if we should proceed.", 1),
    ("You know, it's kind of like, complicated.", 1),
    ("Um, maybe we could reschedule?", 1),
    ("So, I was like, totally confused.", 1),
    ("Well, I mean, it depends, you know?", 1),
    ("Basically, it's not that simple.", 1),
    ("Like, I guess we could try again.", 1),
    ("I mean, you know how it is.", 1),
    ("Uh, that's what I was thinking too.", 1),
    ("So, actually, we should wait.", 1),

    # Not filler (0)
    ("Let's finalize the report by Friday.", 0),
    ("The budget was approved last week.", 0),
    ("Our next meeting is scheduled for Monday.", 0),
    ("We exceeded our sales targets this quarter.", 0),
    ("Please review the latest draft.", 0),
    ("The system was updated overnight.", 0),
    ("Thank you all for your hard work.", 0),
    ("We'll need to update the client.", 0),
    ("The project deadline remains unchanged.", 0),
    ("The final report is ready for submission.", 0)
]


# Loop through each model and evaluate
for model_name, model in models.items():
    print(f"\n Evaluating Model: {model_name}")
    correct = 0
    total = len(test_sentences)

    for sentence, true_label in test_sentences:
        pred_label, prob = predict_sentence(sentence, model)
        is_correct = pred_label == true_label
        correct += is_correct

        print(f"\n Sentence: {sentence}")
        print(f" Expected: {true_label},  Predicted: {pred_label},  Correct: {is_correct}")
        print(f" Probabilities: {prob}")

    accuracy = correct / total
    print(f"\n {model_name} Accuracy on Test Set: {accuracy:.2%}")
```
